```python
def one_hot_encode(numbers, max_num):
    encoded = [0] * (max_num)  # Create a list of zeros with length max_num + 1
    for number in numbers:
        encoded[number-1] = 1  # Set the corresponding index to 1
    return encoded
```

```python
def list_one_hot(df):

    full=[]
    max_num = int(df["modality"][0])
      
    for index in range(len(df)):
        list_= one_hot_encode(ast.literal_eval(df["frames"][index]), max_num)
        full.extend(list_)

    return full
```

```python
def plot_stat (cm, name_):
    
    disp=sns.heatmap(cm,  cmap='Blues', annot=True,  yticklabels=["no peaks", "peaks"], 
                     xticklabels=["no peaks", "peaks"])

    # Set the Title
    disp.set(title="Confusion Matrix")

    # Set the Labels
    disp.set(xlabel=name_, ylabel="Agreement")
  
    disp.plot()
    plt.savefig(f"res/{name_}")
    plt.show()  
```

```python

def save_stat (stats, header, name_file) : 
    with open(f'res/{name_file}.csv', 'w', newline='') as file:
        writer = csv.writer(file) 
        
        writer.writerow(header)
        
        # write a row to the csv file
        writer.writerow(stats)
```

```python
def score_agreement(csv_path_1, csv_path_2, name_=None):
        
    
    df_1 = pd.read_csv(csv_path_1)
    df_2 = pd.read_csv(csv_path_2)
    
    if (list(df_1["uuid"]) != list(df_2["uuid"])):
        raise Exception("uuid are not matching")
        
    l_consensus = list_one_hot(df_1)
    l_perso = list_one_hot(df_2)
    
   
    kappas = cohen_kappa_score(l_consensus, l_perso)
    agrees = (np.mean([int(l_consensus[i]==l_perso[i]) for i in range(len(l_consensus))]))
    stat_cm=classification_report(l_consensus, l_perso)
    cm =confusion_matrix(l_consensus, l_perso)
    save_stat([kappas, agrees, stat_cm], [f"Kappas_{name_}", f"Agree_{name_}", "stat"], "p")
    plot_stat(cm, name_)
    
    

    return kappas,agrees, stat_cm, cm
```

```python
from sklearn.metrics import confusion_matrix

def create_confusion_matrix(csv_path_1, csv_path_2):
    df_1 = pd.read_csv(csv_path_1)
    df_2 = pd.read_csv(csv_path_2)
    merged_df = pd.merge(df_1, df_2, on="uuid")
    max_num = int(merged_df["modality_y"][0])
    
    confusion_matrix_list = []
    for index, row in merged_df.iterrows():
        list_1 = one_hot_encode(ast.literal_eval(row["frames_x"]), max_num)
        list_2 = one_hot_encode(ast.literal_eval(row["frames_y"]), max_num)
        confusion_matrix_list.append(confusion_matrix(list_1, list_2))
    return confusion_matrix_list
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def make_confusion_matrix(cf,
                          group_names=None,
                          categories='auto',
                          count=True,
                          percent=True,
                          cbar=True,
                          xyticks=True,
                          xyplotlabels=True,
                          sum_stats=True,
                          figsize=None,
                          cmap='Blues',
                          title=None):
    '''
    This function will make a pretty plot of an sklearn Confusion Matrix cm using a Seaborn heatmap visualization.

    Arguments
    ---------
    cf:            confusion matrix to be passed in

    group_names:   List of strings that represent the labels row by row to be shown in each square.

    categories:    List of strings containing the categories to be displayed on the x,y axis. Default is 'auto'

    count:         If True, show the raw number in the confusion matrix. Default is True.

    normalize:     If True, show the proportions for each category. Default is True.

    cbar:          If True, show the color bar. The cbar values are based off the values in the confusion matrix.
                   Default is True.

    xyticks:       If True, show x and y ticks. Default is True.

    xyplotlabels:  If True, show 'True Label' and 'Predicted Label' on the figure. Default is True.

    sum_stats:     If True, display summary statistics below the figure. Default is True.

    figsize:       Tuple representing the figure size. Default will be the matplotlib rcParams value.

    cmap:          Colormap of the values displayed from matplotlib.pyplot.cm. Default is 'Blues'
                   See http://matplotlib.org/examples/color/colormaps_reference.html
                   
    title:         Title for the heatmap. Default is None.

    '''


    # CODE TO GENERATE TEXT INSIDE EACH SQUARE
    blanks = ['' for i in range(cf.size)]

    if group_names and len(group_names)==cf.size:
        group_labels = ["{}\n".format(value) for value in group_names]
    else:
        group_labels = blanks

    if count:
        group_counts = ["{0:0.0f}\n".format(value) for value in cf.flatten()]
    else:
        group_counts = blanks

    if percent:
        group_percentages = ["{0:.2%}".format(value) for value in cf.flatten()/np.sum(cf)]
    else:
        group_percentages = blanks

    box_labels = [f"{v1}{v2}{v3}".strip() for v1, v2, v3 in zip(group_labels,group_counts,group_percentages)]
    box_labels = np.asarray(box_labels).reshape(cf.shape[0],cf.shape[1])


    # CODE TO GENERATE SUMMARY STATISTICS & TEXT FOR SUMMARY STATS
    if sum_stats:
        #Accuracy is sum of diagonal divided by total observations
        accuracy  = np.trace(cf) / float(np.sum(cf))

        #if it is a binary confusion matrix, show some more stats
        if len(cf)==2:
            #Metrics for Binary Confusion Matrices
            precision = cf[1,1] / sum(cf[:,1])
            recall    = cf[1,1] / sum(cf[1,:])
            f1_score  = 2*precision*recall / (precision + recall)
            stats_text = "\n\nAccuracy={:0.3f}\nPrecision={:0.3f}\nRecall={:0.3f}\nF1 Score={:0.3f}".format(
                accuracy,precision,recall,f1_score)
        else:
            stats_text = "\n\nAccuracy={:0.3f}".format(accuracy)
    else:
        stats_text = ""


    # SET FIGURE PARAMETERS ACCORDING TO OTHER ARGUMENTS
    if figsize==None:
        #Get default figure size if not set
        figsize = plt.rcParams.get('figure.figsize')

    if xyticks==False:
        #Do not show categories if xyticks is False
        categories=False


    # MAKE THE HEATMAP VISUALIZATION
    plt.figure(figsize=figsize)
    sns.heatmap(cf,annot=box_labels,fmt="",cmap=cmap,cbar=cbar,xticklabels=categories,yticklabels=categories)

    if xyplotlabels:
        plt.ylabel('True label')
        plt.xlabel('Predicted label' + stats_text)
    else:
        plt.xlabel(stats_text)
    
    if title:
        plt.title(title)
```

```python
labels = ['True Neg','False Pos','False Neg','True Pos']
categories = ['No peak', 'peak']
for confusion_matrix in confusion_matrix_list:
    make_confusion_matrix(confusion_matrix, 
                          group_names=labels,
                          categories=categories, 
                          cmap='binary')
```

```python
from quica.quica import Quica
import pandas as pd

def get_quica_results(csv_path_1, csv_path_2):
    df_1 = pd.read_csv(csv_path_1)
    df_2 = pd.read_csv(csv_path_2)
    merged_df = pd.merge(df_1, df_2, on="uuid")
    max_num = int(merged_df["modality_y"][0])
    
    quica_results_list = []
    for index, row in merged_df.iterrows():
        list_1 = one_hot_encode(ast.literal_eval(row["frames_x"]), max_num)
        list_2 = one_hot_encode(ast.literal_eval(row["frames_y"]), max_num)
        if len(set(list_1)) and len(set(list_2)) > 1:
            dataframe = pd.DataFrame({"Laurent" : list_1, "Flavie" : list_2})
            quica_results_list.append(Quica(dataframe=dataframe))
    return quica_results_list
```

```python
quica_results = get_quica_results('./Annotation_20_samples/OCT_annotation_Laurent/annotation_320.csv', './Annotation_20_samples/OCT_annotation_Flavie/annotation_320.csv')
print(quica_results[0].get_results())
for quica in quica_results:
    print(quica.get_results())
```
